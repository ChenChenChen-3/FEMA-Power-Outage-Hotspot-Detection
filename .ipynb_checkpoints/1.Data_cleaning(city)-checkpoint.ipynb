```python
import pandas as pd
```

```python
df = pd.read_csv('./US_City_Data/USCFinal.csv', index_col=0)

# Filter out the data with the states of New England.
df = df[df['state_id'].isin(["ME", "VT", "NH","MA", "CT", "RI"])]
df.head()
```

```python
# Check and drop meaningless columns. 
df.info()
```

```python
df = df.drop(['city_ascii', 
              'state_name', 
              'county_fips', 
              'population', 
              'population_proper', 
              'source', 
              'incorporated'], axis=1)
df.head()
```

```python
# Save cleaned city data.
df.to_csv('./US_City_Data/USCFinal_cleaned.csv', index=False)
```
